* 1c (30pt) Set up the algorithm and plot the positions of the Earth and Jupiter using the Velocity Verlet algorithm. Discuss the stability of the solutions using your Velocity Verlet solver.

```python
# I'm going to define a function which takes as parameters:
# the gravitational constant G
# the number of particles, 
# the initial positions and velocities of the particles in numpy matrix form,
# a time array, and
# a list of the masses of the particles.
# The output will be two tensors: one of the positions, the other of the velocities.

import numpy as np

def G_2D_vv(G,N,M,t,init_r,init_v):
    t_l = len(t)
    r = np.zeros((t_l,N,2))
    r[0] = init_r
    v = np.zeros((t_l,N,2))
    v[0] = init_v
    a = np.zeros((N,N,2))
    for i in range(t_l-1):
        t_s = t[i+1]-t[i]
        
        if i == 0:
            for mu in range(N):
                for nu in range(mu+1,N):
                    a[mu][nu] = - G * M[nu] * ( r[i][mu] - r[i][nu] ) / np.linalg.norm(r[i][mu] - r[i][nu])**3
                    a[nu][mu] = - a[mu][nu] * M[mu]/M[nu]
            a_old = np.sum(a,axis=1)
        else:
            a_old=a_new
            
        r[i+1] = r[i] + v[i] * t_s + 0.5 * a_old * t_s**2
        
        for mu in range(N):
            for nu in range(mu+1,N):
                a[mu][nu] = - G * M[nu] * ( r[i+1][mu] - r[i+1][nu] ) / np.linalg.norm(r[i+1][mu] - r[i+1][nu])**3
                a[nu][mu] = - a[mu][nu] * M[mu]/M[nu]
        a_new = np.sum(a,axis=1)
        
        v[i+1] = v[i] + 0.5 * (a_old+a_new) * t_s
    
    return r, v

G = 1.2 * 10**(-4)

N = 3

M = [3.3*10**5,1,3.2*10**2]

t = np.arange(0,100,0.01)

r_init = [[0,0],[1,0],[5.2,0]]

# We also define a correction velocity since in the code, we allow the sun to move 
# and we don't want the system to be moving away

v_correction = - 2*np.pi/5.2**(0.5) * 10**(-3)

v_init = [[0,v_correction],[0,2*np.pi+v_correction],[0,2*np.pi/5.2**(0.5)+v_correction]]

r, v = G_2D_vv(G,N,M,t,r_init,v_init)

import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots(1,2,figsize=(20,10))

ax[0].plot(r[:,0,0],r[:,0,1],label='Sun')
ax[0].plot(r[:,1,0],r[:,1,1],label='Earth')
ax[0].plot(r[:,2,0],r[:,2,1],label='Jupiter')
ax[0].set_title('Orbital motion')
ax[0].set_xlabel('x [AU]')
ax[0].set_ylabel('y [AU]')

ax[1].plot(t,v[:,0,1],label='Sun')
ax[1].plot(t,v[:,1,1],label='Earth')
ax[1].plot(t,v[:,2,1],label='Jupiter')
ax[1].set_title('Velocity in x-direction vs time')
ax[1].set_xlabel('t [yr]')
ax[1].set_ylabel('v_x [AU/yr]')

ax[0].grid()
ax[0].legend()
ax[1].grid()
ax[1].legend()
```

As we see, the system seems to be stable and the motion is circular as we predict. Let's also write some other functions to calculate the energy so that we can plot energy over time.

```python
def K(v,M):
    t_l = v.shape[0]
    N = len(M)
    K = []
    for i in range(t_l):
        K.append(0)
        for mu in range(N):
            K[i] += 0.5 * M[mu] * np.linalg.norm(v[i][mu])**2
    return np.array(K)
            
def V(G,v,M):
    t_l = r.shape[0]
    N = len(M)
    V = []
    for i in range(t_l):
        V.append(0)
        for mu in range(N)
            for nu in range(mu+1,N):
                V[i] += - G * M[mu]*M[nu] / np.linalg.norm(r[i][mu]-r[i][nu])
    return np.array(V)

fig, ax = plt.subplots(1,1,figsize=(12,6))
ax.plot(t,K(v,M)+V(G,v,M),label='Total')
ax.plot(t,K(v,M),label='Kinetic')
ax.plot(t,V(G,v,M),label='Potential')
ax.set_title('Energy vs time')
ax.set_xlabel('t [yr]')
ax.legend()
ax.set_ylabel(r'E $\left[\frac{M_E AU^2}{yr^2}\right]$')
ax.grid()
```

Hence, the motion seems to be stable and the energy seems to be conserved.

An important test of the general theory of relativity was to compare
its prediction for the perihelion precession of Mercury to the
observed value. The observed value of the perihelion precession, when
all classical effects (such as the perturbation of the orbit due to
gravitational attraction from the other planets) are subtracted, is
$43''$ ($43$ arc seconds) per century.

Closed elliptical orbits are a special feature of the Newtonian
$1/r^2$ force. In general, any correction to the pure $1/r^2$
behaviour will lead to an orbit which is not closed, i.e. after one
complete orbit around the Sun, the planet will not be at exactly the
same position as it started. If the correction is small, then each
orbit around the Sun will be almost the same as the classical ellipse,
and the orbit can be thought of as an ellipse whose orientation in
space slowly rotates. In other words, the perihelion of the ellipse
slowly precesses around the Sun.

You will now study the orbit of Mercury around the Sun, adding a general relativistic correction to the Newtonian
gravitational force, so that the force becomes

$$
F = -\frac{GM_\mathrm{Sun}M_\mathrm{Mercury}}{r^2}\left[1 + \frac{3l^2}{r^2c^2}\right]
$$

where $M_\mathrm{Mercury}$ is the mass of Mercury, $r$ is the distance
between Mercury and the Sun, $l=|\vec{r}\times\vec{v}|$ is the
magnitude of Mercury's orbital angular momentum per unit mass, and $c$
is the speed of light in vacuum. Run a simulation over one century of
Mercury's orbit around the Sun with no other planets present, starting
with Mercury at the perihelion on the $x$ axis.  Check then the value
of the perihelion angle $\theta_\mathrm{p}$, using

$$
\tan \theta_\mathrm{p} = \frac{y_\mathrm{p}}{x_\mathrm{p}}
$$

where $x_\mathrm{p}$ ($y_\mathrm{p}$) is the $x$ ($y$) position of
Mercury at perihelion, i.e. at the point where Mercury is at its
closest to the Sun. You may use that the speed of Mercury at
perihelion is $12.44\,\mathrm{AU}/\mathrm{yr}$, and that the distance
to the Sun at perihelion is $0.3075\,\mathrm{AU}$.

You need to make
sure that the time resolution used in your simulation is sufficient,
for example by checking that the perihelion precession you get with a
pure Newtonian force is at least a few orders of magnitude smaller
than the observed perihelion precession of Mercury. Can the observed
perihelion precession of Mercury be explained by the general theory of
relativity?

We will define now another function which uses a general relativistic formula for gravity. We also need to write the functions anew and specifically for a two-body problem with the position of the sun fixed since we want our simulation to be as fast as possible to observe the precession accurately. While the equation probably won't be fully accurate given a lack of technical background knowledge on general relativity, it will be approximately correct.

```python
import numpy as np

def Solar_GR_vv(G,c,M,t,r_init,v_init):
    t_l = len(t)
    r = np.zeros((t_l,2))
    r[0] = r_init
    v = np.zeros((t_l,2))
    v[0] = v_init
    l = np.abs(r_init[0] * v_init[1] - r_init[1] * v_init[0])
    r_init_abs = np.linalg.norm(r_init)
    a_new = - G * M * r_init / r_init_abs**3 * (1 + 3 / r_init_abs**2 / c**2 * l**2)
    for i in range(t_l-1):
        
        t_s = t[i+1]-t[i]
        a_old = a_new
        
        r[i+1] = r[i] + v[i] * t_s + 0.5 * a_old * t_s**2
        
        r_abs = np.linalg.norm(r[i+1])
        a_new = - G * M * r[i+1] / r_abs**3 * (1 + 3 / r_abs**2 / c**2 * l**2)

        v[i+1] = v[i] + 0.5 * (a_old+a_new) * t_s
    
    return r, v

def Solar_G_vv(G,M,t,r_init,v_init):
    t_l = len(t)
    r = np.zeros((t_l,2))
    r[0] = r_init
    v = np.zeros((t_l,2))
    v[0] = v_init
    r_init_abs = np.linalg.norm(r_init)
    a_new = - G * M * r_init / r_init_abs**3
    for i in range(t_l-1):
        
        t_s = t[i+1]-t[i]
        a_old = a_new
        
        r[i+1] = r[i] + v[i] * t_s + 0.5 * a_old * t_s**2
        
        r_abs = np.linalg.norm(r[i+1])
        a_new = - G * M * r[i+1] / r_abs**3

        v[i+1] = v[i] + 0.5 * (a_old+a_new) * t_s
    
    return r, v

G = 1.2 * 10**(-4)
c = 6.4 * 10**4 # in units AU/yr
M = 3.3*10**5

t = np.arange(0,100,0.0001)
r_init = np.array([0.3075,0])
v_init = np.array([0,12.44])

r_G, v_G = Solar_G_vv(G,M,t,r_init,v_init)

r_GR, v_GR = Solar_GR_vv(G,c,M,t,r_init,v_init)

fig, ax = plt.subplots(2,2,figsize=(16,16))

ax[0,0].plot(r_G[:,0],r_G[:,1],label='Mercury')
ax[0,0].set_title(r'Orbital motion using Newton\'s law')
ax[0,0].set_xlabel('x [AU]')
ax[0,0].set_ylabel('y [AU]')
ax[0,0].plot([0],[0],label='Sun',marker='o',linestyle='')

ax[0,1].plot(r_GR[:,0],r_GR[:,1],label='Mercury')
ax[0,1].set_title(r'Orbital motion using General Relativity')
ax[0,1].set_xlabel('x [AU]')
ax[0,1].set_ylabel('y [AU]')
ax[0,1].plot([0],[0],label='Sun',marker='o',linestyle='')

ax[1,0].scatter(np.arctan(r_G[:,1]/r_G[:,0]),r_G[:,0]**2 + r_G[:,1]**2,label='Mercury')
ax[1,0].set_title(r'Radius vs angle using Newton\'s law')
ax[1,0].set_ylabel('Radius [AU]')
ax[1,0].set_xlabel('Angle [rad]')

ax[1,1].scatter(np.arctan(r_GR[:,1]/r_GR[:,0]),r_GR[:,0]**2 + r_GR[:,1]**2,label='Mercury')
ax[1,1].set_title(r'Radius vs angle using General Relativity')
ax[1,1].set_ylabel('Radius [AU]')
ax[1,1].set_xlabel('Angle [rad]')
```

Let's now define some runge-kutta solvers to see if we can get more accurate results.

```python
def Solar_G_rk4(G,M,t,r_init,v_init):
    t_l = len(t)
    r = np.zeros((t_l,2))
    r[0] = r_init
    v = np.zeros((t_l,2))
    v[0] = v_init
    for i in range(t_l-1):
        
        ts = t[i+1]-t[i]
        
        rabs = np.linalg.norm(r[i])
        
        k1r = v[i]
        k1v = - G * M * r[i] / rabs**3
        
        rr = r[i] + 0.5 * k1r * ts
        rabs = np.linalg.norm(rr)
        
        k2r = v[i] + 0.5 * k1v * ts
        k2v = - G * M * rr / rabs**3
        
        rr = r[i] + 0.5 * k2r * ts
        rabs = np.linalg.norm(rr)
        
        k3r = v[i] + 0.5 * k2v * ts
        k3v = - G * M * rr / rabs**3
        
        rr = r[i] + k3r * ts
        rabs = np.linalg.norm(rr)
        
        k4r = v[i] + k3v * ts
        k4v = - G * M * rr / rabs**3
        
        r[i+1] = r[i] + (1/6 * k1r + 2/6 * k2r + 2/6 * k3r + 1/6 * k4r) * ts
        v[i+1] = v[i] + (1/6 * k1v + 2/6 * k2v + 2/6 * k3v + 1/6 * k4v) * ts
    
    return r, v

def Solar_GR_rk4(G,c,M,t,r_init,v_init):
    t_l = len(t)
    r = np.zeros((t_l,2))
    r[0] = r_init
    v = np.zeros((t_l,2))
    v[0] = v_init
    l = np.abs(r_init[0] * v_init[1] - r_init[1] * v_init[0])
    eps = 3 / c**2 * l**2
    for i in range(t_l-1):
        
        ts = t[i+1]-t[i]
        
        rabs = np.linalg.norm(r[i])
        
        k1r = v[i]
        k1v = - G * M * r[i] * (1 + eps / rabs**2) / rabs**3
        
        rr = r[i] + 0.5 * k1r * ts
        rabs = np.linalg.norm(rr)
        
        k2r = v[i] + 0.5 * k1v * ts
        k2v = - G * M * rr * (1 + eps / rabs**2) / rabs**3
        
        rr = r[i] + 0.5 * k2r * ts
        rabs = np.linalg.norm(rr)
        
        k3r = v[i] + 0.5 * k2v * ts
        k3v = - G * M * rr * (1 + eps / rabs**2) / rabs**3
        
        rr = r[i] + k3r * ts
        rabs = np.linalg.norm(rr)
        
        k4r = v[i] + k3v * ts
        k4v = - G * M * rr * (1 + eps / rabs**2) / rabs**3
        
        r[i+1] = r[i] + (1/6 * k1r + 2/6 * k2r + 2/6 * k3r + 1/6 * k4r) * ts
        v[i+1] = v[i] + (1/6 * k1v + 2/6 * k2v + 2/6 * k3v + 1/6 * k4v) * ts
    
    return r, v
```

```python
G = 1.2 * 10**(-4)
c = 6.4 * 10**4 # in units AU/yr
M = 3.3 * 10**5

t = np.arange(0,10,0.001)
r_init = np.array([0.3075,0])
v_init = np.array([0,12.44])

r_G, v_G = Solar_G_rk4(G,M,t,r_init,v_init)

r_GR, v_GR = Solar_GR_rk4(G,c,M,t,r_init,v_init)

fig, ax = plt.subplots(2,2,figsize=(16,16))

ax[0,0].plot(r_G[:,0],r_G[:,1],label='Mercury')
ax[0,0].set_title(r'Orbital motion using Newton\'s law')
ax[0,0].set_xlabel('x [AU]')
ax[0,0].set_ylabel('y [AU]')
ax[0,0].plot([0],[0],label='Sun',marker='o',linestyle='')

ax[0,1].plot(r_GR[:,0],r_GR[:,1],label='Mercury')
ax[0,1].set_title(r'Orbital motion using General Relativity')
ax[0,1].set_xlabel('x [AU]')
ax[0,1].set_ylabel('y [AU]')
ax[0,1].plot([0],[0],label='Sun',marker='o',linestyle='')

ax[1,0].scatter(np.arctan(r_G[:,1]/r_G[:,0]),r_G[:,0]**2 + r_G[:,1]**2,label='Mercury')
ax[1,0].set_title(r'Radius vs angle using Newton\'s law')
ax[1,0].set_ylabel('Radius [AU]')
ax[1,0].set_xlabel('Angle [rad]')

ax[1,1].scatter(np.arctan(r_GR[:,1]/r_GR[:,0]),r_GR[:,0]**2 + r_GR[:,1]**2,label='Mercury')
ax[1,1].set_title(r'Radius vs angle using General Relativity')
ax[1,1].set_ylabel('Radius [AU]')
ax[1,1].set_xlabel('Angle [rad]')
```

Now let's define another function to output two arrays: one for the radius and one for the angle.

```python
def polar(r):
    N = r.shape[0]
    angle = np.zeros(N)
    r_abs = np.zeros(N)
    for i in range(N):
        x = r[i][0]
        y = r[i][1]
        r_abs[i]=x**2 + y**2
        if x > 0 and y >= 0:
            angle[i]=np.arctan(y/x)
        elif x > 0 and y <= 0:
            angle[i]=np.arctan(y/x)+2*np.pi
        elif x < 0 and y >= 0:
            angle[i]=np.arctan(y/x)+np.pi
        elif x < 0 and y <= 0:
            angle[i]=np.arctan(y/x)+np.pi
        else:
            angle[i]=np.pi - np.sign(y)*np.pi/2
    return angle, r_abs
```

Since the deviation of the periods is very small, we can't discern it by eye. We will have to use a function fit. We also see that even though we calculated the period to high accuracy (with the given parameters) using a small timestep, the period still diverges for different timesteps simply due to the inaccuracy of the simulation. So we will graph the difference between the two different acceleration expressions versus the timestep.

```python
from scipy.optimize import curve_fit

def ellipse(x, a, b, c, d):
    return a / (1 + b * np.cos(c * x + d))

G = 1.2 * 10**(-4)
c = 6.4 * 10**4 # in units AU/yr
M = 3.3*10**5

r_init = np.array([0.3075,0])
v_init = np.array([0,12.44])
ts_list = [0.01,0.008,0.006,0.005,0.004,0.003,0.002,0.0015,0.001,0.0006,0.0003,0.0002,0.0001]

c_diff_list = []

for ts in ts_list:
    
    t = np.arange(0,100.238692,ts)
    
    i_0 = int(len(t) * 100/100.238692)
    i_1 = int(len(t))
    
    r_GR, v_GR = Solar_GR_vv(G,c,M,t,r_init,v_init)
    r_G, v_G = Solar_G_vv(G,M,t,r_init,v_init)
    
    angle_G, radius_G = polar(r_G)
    angle_GR, radius_GR = polar(r_GR)
    
    p_GR, p_GR_2 = curve_fit(ellipse, angle_GR[i_0:i_1], radius_GR[i_0:i_1],p0=(0.2,0.2,1,0))
    p_G, p_G_2 = curve_fit(ellipse, angle_G[i_0:i_1], radius_G[i_0:i_1],p0=(0.2,0.2,1,0))
    
    c_diff_list.append(np.abs(p_GR[3]-p_G[3]))
    
# But c_diff_list contains the deviations in units of radians
# Let's convert that to arcseconds

arc_list = []

for i in range(len(c_diff_list)):
    c = c_diff_list[i]
    arc = c * 60 * 60 * 180 / np.pi
    arc_list.append(arc)
    
# Let's now plot the results
    
fig, ax = plt.subplots(1,1,figsize=(16,8))

ax.plot(ts_list,arc_list,linestyle='--',marker='^')
ax.grid()
ax.set_xlabel('timestep [yr]')
ax.set_ylabel('angle deviation [arcs]')
ax.set_title('angle deviation vs timestep')
```

From the final graph, we see that we do have a difference between the orbit deviations which converges to a certain value. That cannot be explained in terms of the inaccuracy of the simulation. This deviation therefore has to come from the difference of adding the extra term to the formula for gravitational attraction which comes from general relativity. Therefore, we can see the periphelion of Mercury process. And the value of that deviation over a century seems to be somewhere close to $40$ arcseconds (with an inaccuracty of $\pm 10$ arcseconds). This is because a line of best fit through this graph seems like it would cross somewhere around $40$ arcseconds. However, a precise value for this deviation is hard to find exactly.
